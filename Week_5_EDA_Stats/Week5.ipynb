```python
import scipy.stats as stats
from math import sqrt
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

### Example 1
S. Belirli bir sosyal yardım alıcı grubu, 20 dolar standart  sapmayla haftada 110 $ SNAP yardımı alıyor. 25 kişiden oluşan rastgele bir örnek alınırsa, ortalama yardimlarin haftada 120 dolardan fazla olma olasılığı nedir?

```python
x_bar = 120 # sample mean
n = 25 # number of samples
sigma = 20 # sd of population
mu = 110 # Population mean
```

```python
z = (x_bar - mu)/ (sigma/sqrt(n))
z
```

Step 2: Look up the z-score in a table (or calculate it using technology).
A z-score of 2.5 has an area of roughly 49.38%. Adding 50% (for the left half of the curve), we get 99.38%.

```python
p_value = 1 - stats.norm.cdf(z)
p_value
```

```python
# Create a range of values for the x-axis
x = np.linspace(-4, 4, 1000)

# Plot the standard normal distribution curve
plt.plot(x, norm.pdf(x, 0, 1), 'r-', lw=2)

# Shade the area to the right of the z-value
plt.fill_between(x[x > z], norm.pdf(x[x > z], 0, 1), color='blue', alpha=0.5)

# Mark the z-value on the x-axis
plt.axvline(x=z, color='k', linestyle='--')

# Add labels and title
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution with z = 2.5')
plt.text(z - 0.5, 0.1, 'p-value ≈ {:.4f}'.format(p_value), fontsize=10)
# Display the plot
plt.show()
```

### Example 2.
29 yaşındaki erkeklerin ortalama maaşı 29.321 dolar, standart sapması ise 2.120 dolar. 100 erkekten oluşan bir örnek alınırsa, ortalama maaşlarının 29.000 dolardan az olma olasılığı nedir?

```python
#H0 = 
#Ha =
```

```python
x_bar = 29.0 # sample mean
n = 100 # number of samples
sigma = 2.120 # sd of population
mu = 29.321 # Population mean
```

```python
z = (x_bar - mu)/ (sigma/sqrt(n))
z
```

Step 2: Look up the z-score in the left-hand z-table (or use technology). -1.51 has an area of 93.45%.

However, this is not the answer, as the question is asking for LESS THAN, and 93.45% is the area “greater than” so you need to subtract from 100%.
100% – 93.45% = 6.55% or about 0.07.

```python
p_value = stats.norm.cdf(z)
p_value
```

```python
# Create a range of values for the x-axis
x = np.linspace(-4, 4, 1000)

# Plot the standard normal distribution curve
plt.plot(x, stats.norm.pdf(x, 0, 1), 'r-', lw=2)

# Shade the area corresponding to the p-value
plt.fill_between(x[x <= z], stats.norm.pdf(x[x <= z], 0, 1), color='blue', alpha=0.5)

# Mark the z-value on the x-axis
plt.axvline(x=z, color='k', linestyle='--')

# Add labels and title
plt.xlabel('Z')
plt.ylabel('Probability Density')
plt.title('Standard Normal Distribution with z = -1.5141')

# Show p-value
plt.text(z - 0.5, 0.1, 'p-value ≈ {:.4f}'.format(p_value), fontsize=10)

# Display the plot
plt.show()
```
